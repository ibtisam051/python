## Student Management System

```python
student_name=[]
student_rol=[]
while True:
    print('''
            1. Add Students
            2. List all students.
            3. Search students
            4. Remove Students
            5. Quit
    
    ''')
    option=int(input("Enter number for the menu from 1 to 5: "))
    if option==1:
        name=input("Enter the name of student : ")
        roll=int(input("Enter the roll number of student : "))
        student_name.append(name)
        student_rol.append(roll)
    elif option==2:
        print("Roll Number : ",student_rol)
        print("Student name : ",student_name)
    elif option==3:
        search_roll=int(input("Enter the roll number of that student : "))
        for i in student_rol:
            if search_roll in student_rol:
                print("roll number :",search_roll,"\n found")
            else:
                print("not found")     
    elif option==4:
        search_name=input("Enter the name of student you want to search : ")
        search_roll=int(input("Enter the roll number of that student : "))
        if (search_name in student_name)and (search_roll in student_rol):
            student_name.remove(search_name)
            student_rol.remove(search_roll)
            print("removed")
        else:
          print("not found") 
    elif option==5:
        print("program terminated!!")
        break
    else:
        print("Invalid!!")
```
